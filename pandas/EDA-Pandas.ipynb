# Exploratory Data Analysis: Pandas -  (Part 1)

#### Import useful library

```python
import pandas as pd
```

#### Import Data

```python
# Load the dataset
iris_data = pd.read_csv('./data/iris.csv')
```

#### Basic Structures

```python
# use .info to get brief information about the dataframe
iris_data.info()
```

```python
# use .head() to get the first 5 rows 
iris_data.head()
#iris_data.head(7) # to get the first 7 rows 
```

```python
# use .tail() to get the last 5 rows
iris_data.tail()
#iris_data.tail(7) # # to get the last 7 rows 
```

```python
# column selection using dot (.)
iris_data.sepal_width
```

```python
# selection using column name as string
iris_data['sepal_width']
```

```python
# selecting using multiple culumn name as list of strings
iris_columns = ['sepal_width', 'petal_length']
iris_data[iris_columns]
```

```python
# indexing: use loc for label based indexing
iris_data.loc[5:10,]
```

```python
# selecting specific columns
iris_data.loc[5:10, iris_columns]
```

```python
# indexing (iloc) does not include the last index, they are zero-based
iris_data.iloc[5:10, 1:3]
```

```python
# filter row, based on condition
setosa_iris = iris_data.loc[iris_data.species == 'Iris-setosa']
setosa_iris
#print(len(setosa_iris))
```
